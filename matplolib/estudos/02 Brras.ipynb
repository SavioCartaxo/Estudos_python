```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Gr√°fico em barras

x = np.array(["A", "B", "C", "D"])
y = np.array([8, 3, 5, 2])

plt.bar(x,y ,width=0.2)
plt.show()
```

```python
# Gr√°fico em barras de lado

x = np.array(["A", "B", "C", "D"])
y = np.array([8, 3, 5, 2])
y = np.sort(y)

plt.barh(x,y, color='r', height = 0.1)
plt.show()
```
